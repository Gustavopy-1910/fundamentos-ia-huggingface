```


from transformers import pipeline


qa_model = pipeline("question-answering", model="deepset/roberta-base-squad2")

contexto = """
O Hugging Face é uma empresa americana que desenvolve ferramentas para criar aplicações
usando aprendizado de máquina. É mais conhecida por sua biblioteca Transformers,
que permite que usuários compartilhem modelos e conjuntos de dados de IA.
"""


pergunta = "O que o Hugging Face permite que os usuários compartilhem?"

resultado = qa_model(question=pergunta, context=contexto)

print(f"Pergunta: {pergunta}")
print(f"Resposta da IA: {resultado['answer']}")
```
